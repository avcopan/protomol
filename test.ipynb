```python
A_ = "Cv4,Ov2"
Ar = "Cv3,Ov1"


class ReactionSmarts:
    """SMARTS reaction templates for enumeration."""

    abstraction = f"[{A_}:1][H:2].[{Ar}:3]>>[{Ar}:1].[H:2][{A_}:3]"
    h_migration = f"([{Ar}:1].[{A_}:2][H:3])>>([H:3][{A_}:1].[{Ar}:2])"
```

```python
from protomol.rd import mols as m_
```

```python
from rdkit.Chem import Draw

reactants = m_.from_smiles("CCC", "[OH]")
products = m_.from_smiles("CC[CH2]", "O")
smarts = ReactionSmarts.abstraction

display(Draw.MolsToImage(m_.with_flat_index_numbers(reactants)))
display(Draw.MolsToImage(m_.with_flat_index_numbers(products)))

print("Reactant with atoms mapped onto product:")
mapping = m_.reaction_mapping(reactants, products, smarts=smarts)
if mapping is None:
    print("No mapping...")
else:
    display(Draw.MolsToImage(m_.with_flat_index_numbers(reactants, mapping)))

print(mapping)
```
